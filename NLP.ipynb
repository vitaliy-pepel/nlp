```python
def tokenize_(text):
    text = re.sub(r'[^\w\s]' , '', text.lower())
    tokens = word_tokenize(text, language="russian")
    return tokens

def stopwords_del_and_lemmatize(tokens):
    _stopwords = stopwords.words("russian")
    filtered_tokens = [token for token in tokens if token not in _stopwords]
    lemmatizer =  pymystem3.Mystem()
    lemm_filtered_tokens = [lemmatizer.lemmatize(token)[0] for token in filtered_tokens]
    return lemm_filtered_tokens
```
